```python
#BeautifulSoup is used for webscraping
from urllib.request import urlopen
html=urlopen('https://www.hubertiming.com/results/2018MLK')
html
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html)
title=soup.title
print(title.text)
```

```python
data=[]
allrows=soup.find_all('tr')
for row in allrows:
    allcols=row.find_all('td')
    #print(allcols)
    datarow=[]
    for col in allcols:
        datarow.append(col.text)
    data.append(datarow)
print(data[5:])
    
```

```python
import pandas as pd
dataframe=pd.DataFrame(data[5:])
dataframe
```

```python
#Logic to get the column names
column=[]
header=soup.find_all('th')
for h in header:
    column.append(h.text)
column
dataframe.columns=column
dataframe
```
